```python


from sklearn.impute import SimpleImputer
import numpy as np

#array with missing value
Marks = [[80],[90],["NaN"], [70],[89],[25]]

#setting up structure of array with numpy 
build = np.array([[80,],[90],[np.nan],[70],[89],[25]]) 

#using simpleimputer for filling missing value with the use of median 
imputer = SimpleImputer(strategy="median")

#fit_transform calculate the strategy -(mean,median,most_frequency) and replace the nan values in the given array (Marks)
Clean_data = imputer.fit_transform (Marks) 
print(Clean_data)
```
