```python
import mysql.connector
try:
  conn = mysql.connector.connect (
    host ="localhost",
    user='root',
    password="password",
    database="model_1"
    )
  if conn.is_connected():
        print("Successfully connected to database")

except mysql.connector.Error as err:
    print(f"Error :{err}")

cursor = conn.cursor()


cursor.close()
conn.close()
print("MySQL connection is closed")
```

```python
 # fasterst way of inserting dataset into database 

import pandas as pd 
from  sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:password@localhost/model_1")

df = pd.read_excel("test_Data.xlsx")

df.to_sql(
    name='loan_data',
    con = engine ,
    if_exists = 'append',
    index = False,
    chunksize = 500 ,
    method ='multi'
)
print("Data uploaded successfully!")
```

```python
import pandas as pd 
from sqlalchemy import create_engine


engine = create_engine("mysql+mysqlconnector://root:password@localhost/model_1")


query = """ Select loanStatus from loan_data;"""
query1 = """ select * from loan_data where SEX = 'M';"""


df_q = pd.read_sql(query ,engine) 
print(df_q.head())         # printing first 5 values of loan Status from dataset 

df_q1 = pd.read_sql(query1 , engine)
print(df_q1.tail())          # data for male employee only 

```

```python
# details of clients whith age > 30  ex( > , < ,= ,!= ..etc)
query2 = """  select * from loan_data where AGE > 30;"""
df_q2 = pd.read_sql(query2 , engine)
print(df_q2)


# use of AND on dataset , ex( OR , NOT , BETWEEN n1 AND n2  )
print("    Age and Loan status together :  \n")
query3 = """ select * from loan_data where AGE > 30 AND loanStatus = 'Active';"""

# use of Count 
query4 = """ select COUNT(*) from loan_data where AGE > 30 AND loanStatus = 'Active';"""

df_q3 = pd.read_sql( query3 , engine )
df_q4 = pd.read_sql(query4 , engine)

print(df_q3)
print("Count of data set :")
print(df_q4)

```
