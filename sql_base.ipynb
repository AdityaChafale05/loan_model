```python
 # fasterst way of inserting dataset into database 

import pandas as pd 
from  sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:password@localhost/model_1")

df = pd.read_excel("test_Data.xlsx")

df.to_sql(
    name='loan_data',
    con = engine ,
    if_exists = 'append',
    index = False,
    chunksize = 500 ,
    method ='multi'
)
print("Data uploaded successfully!")
```
