```python
import mysql.connector
try:
  conn = mysql.connector.connect (
    host ="localhost",
    user='root',
    password="password",
    database="model_1"
    )
  if conn.is_connected():
        print("Successfully connected to database")

except mysql.connector.Error as err:
    print(f"Error :{err}")

cursor = conn.cursor()


cursor.close()
conn.close()
print("MySQL connection is closed")
```

```python
import pandas as pd 
from  sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:password@localhost/model_1")

df = pd.read_excel("test_Data.xlsx")

df.to_sql(
    name='loan_data',
    con = engine ,
    if_exists = 'append',
    index = False,
    chunksize = 500 ,
    method ='multi'
)
print("Data uploaded successfully!")
```
